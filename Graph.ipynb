```python
import pandas as pd
import numpy as np
import networkx as nx
df = pd.read_csv ('data2.csv')
adj=np.asmatrix(df)
adj
```

```python
G = nx.DiGraph(adj)
nx.draw_networkx(G, with_labels=True)
sorted(G)
```

```python
mapping={0:"Intesa", 1:"Mediobanca", 2:"Credit Agricole", 3:"Societe Generale", 4:"Banco Santander", 5:"Deutsche Bank", 6:"Commerzbank", 7:"HSBC", 8:"Lloyds bank", 9:"BNP", 10: "Credit Suisse", 11:"UBS", 12:"DanskeBank", 13:"Swedbank"}
H=nx.relabel_nodes(G, mapping, copy=True)
nx.draw_networkx(H, with_labels=True)
```
